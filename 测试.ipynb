```python
from pyecharts import options as opts
from pyecharts.charts import Line, Bar, Grid, Page, Scatter,Page
import pandas as pd
# from pyecharts.charts import Line, 
# 假设 df 是你的 DataFrame，包含以下列：
# "Date", "RSI", "K", "D", "J", "Percent_K", "Percent_D", "DIF", "DEA", "Histogram"
# 模拟数据
from sqlalchemy import create_engine

engine = create_engine("mysql+pymysql://root:chen@127.0.0.1:3306/gp")
df=pd.read_sql("select * from gp.stock_data where code='601360' order by `Date` desc limit 100" ,con=engine)
df=df.sort_values('Date')
# 设置滑动窗口范围
start_index = 0
end_index = 100

# ---------------- RSI 图表 ----------------
# def create_rsi_chart(df):
#     line = (
#         Line()
#         .add_xaxis(df["Date"][start_index:end_index].tolist())
#         .add_yaxis("RSI", df["RSI"][start_index:end_index].tolist(), is_smooth=True, label_opts=opts.LabelOpts(is_show=False))
#         .set_global_opts(
#             title_opts=opts.TitleOpts(title="RSI 图表"),
#             xaxis_opts=opts.AxisOpts(type_="category", boundary_gap=False, axislabel_opts=opts.LabelOpts(rotate=45)),
#             yaxis_opts=opts.AxisOpts(name="RSI 值", max_=100, min_=0),
#             datazoom_opts=[opts.DataZoomOpts(), opts.DataZoomOpts(type_="inside")],
#         )
#     )
#     return line
def create_rsi_chart(df, start_index=0, end_index=None):
    last_index = end_index if end_index is not None else len(df)

    line = (
        Line()
        .add_xaxis(df["Date"][start_index:last_index].tolist())
        .add_yaxis(
            "RSI",
            df["RSI"][start_index:last_index].tolist(),
            is_smooth=True,
            label_opts=opts.LabelOpts(is_show=False),
        )
        .set_global_opts(
            title_opts=opts.TitleOpts(title="RSI 图表"),
            xaxis_opts=opts.AxisOpts(type_="category", boundary_gap=False, axislabel_opts=opts.LabelOpts(rotate=45)),
            yaxis_opts=opts.AxisOpts(name="RSI 值", max_=100, min_=0),
            datazoom_opts=[opts.DataZoomOpts(), opts.DataZoomOpts(type_="inside")],
            tooltip_opts=opts.TooltipOpts(trigger="axis", axis_pointer_type="cross")
        )
    )
    return line
# ---------------- KDJ 图表 ----------------
def create_kdj_chart(df):
    line = (
        Line()
        .add_xaxis(df["Date"][start_index:end_index].tolist())
        .add_yaxis("K", df["K"][start_index:end_index].tolist(), label_opts=opts.LabelOpts(is_show=False))
        .add_yaxis("D", df["D"][start_index:end_index].tolist(), label_opts=opts.LabelOpts(is_show=False))
        .add_yaxis("J", df["J"][start_index:end_index].tolist(), label_opts=opts.LabelOpts(is_show=False))
        .set_global_opts(
            title_opts=opts.TitleOpts(title="KDJ 图表"),
            xaxis_opts=opts.AxisOpts(type_="category", boundary_gap=False, axislabel_opts=opts.LabelOpts(rotate=45)),
            yaxis_opts=opts.AxisOpts(name="KDJ 值"),
            datazoom_opts=[opts.DataZoomOpts(), opts.DataZoomOpts(type_="inside")],
        )
    )
    return line
def create_kdj_chart(df, start_index=0, end_index=None):
    last_index = end_index if end_index is not None else len(df)

    line = (
        Line()
        .add_xaxis(df["Date"][start_index:last_index].tolist())
        .add_yaxis("K", df["K"][start_index:last_index].tolist(), label_opts=opts.LabelOpts(is_show=False))
        .add_yaxis("D", df["D"][start_index:last_index].tolist(), label_opts=opts.LabelOpts(is_show=False))
        .add_yaxis("J", df["J"][start_index:last_index].tolist(), label_opts=opts.LabelOpts(is_show=False))
        .set_global_opts(
            title_opts=opts.TitleOpts(title="KDJ 图表"),
            xaxis_opts=opts.AxisOpts(type_="category", boundary_gap=False, axislabel_opts=opts.LabelOpts(rotate=45)),
            yaxis_opts=opts.AxisOpts(name="KDJ 值"),
            datazoom_opts=[opts.DataZoomOpts(), opts.DataZoomOpts(type_="inside")],
            tooltip_opts=opts.TooltipOpts(trigger="axis", axis_pointer_type="cross")
        )
    )
    return line
# ---------------- LK 图表 ----------------
def create_lk_chart(df, start_index=0, end_index=None):
    last_index = end_index if end_index is not None else len(df)

    bar = (
        Bar()
        .add_xaxis(df["Date"][start_index:last_index].tolist())
        .add_yaxis("Percent_K", df["Percent_K"][start_index:last_index].tolist())
        .set_global_opts(
            title_opts=opts.TitleOpts(title="LK 图表"),
            xaxis_opts=opts.AxisOpts(type_="category"),
            yaxis_opts=opts.AxisOpts(name="LK 值"),
            datazoom_opts=[opts.DataZoomOpts(), opts.DataZoomOpts(type_="inside")],
            tooltip_opts=opts.TooltipOpts(trigger="axis", axis_pointer_type="cross")
        )
    )
    line = (
        Line()
        .add_xaxis(df["Date"][start_index:last_index].tolist())
        .add_yaxis("Percent_K", df["Percent_K"][start_index:last_index].tolist(), label_opts=opts.LabelOpts(is_show=False))
        .add_yaxis("Percent_D", df["Percent_D"][start_index:last_index].tolist(), label_opts=opts.LabelOpts(is_show=False))
    )
    bar.overlap(line)
    return bar

# ---------------- MACD 图表 ----------------
def create_macd_chart(df, start_index=0, end_index=None):
    last_index = end_index if end_index is not None else len(df)

    bar = (
        Bar()
        .add_xaxis(df["Date"][start_index:last_index].tolist())
        .add_yaxis("柱状图", df["Histogram"][start_index:last_index].tolist(), category_gap="50%")
        .set_global_opts(
            title_opts=opts.TitleOpts(title="MACD 图表"),
            xaxis_opts=opts.AxisOpts(type_="category"),
            yaxis_opts=opts.AxisOpts(name="MACD 值"),
            datazoom_opts=[opts.DataZoomOpts(), opts.DataZoomOpts(type_="inside")],
            tooltip_opts=opts.TooltipOpts(trigger="axis", axis_pointer_type="cross")
        )
    )
    line = (
        Line()
        .add_xaxis(df["Date"][start_index:last_index].tolist())
        .add_yaxis("DIF", df["DIF"][start_index:last_index].tolist(), label_opts=opts.LabelOpts(is_show=False))
        .add_yaxis("DEA", df["DEA"][start_index:last_index].tolist(), label_opts=opts.LabelOpts(is_show=False))
    )
    bar.overlap(line)
    return bar

# ---------------- 组合 2 行 2 列的布局 ----------------
def save_charts_to_html(df, output_file="charts_grid.html"):
    # 创建单独图表
    rsi_chart = create_rsi_chart(df)
    kdj_chart = create_kdj_chart(df)
    lk_chart = create_lk_chart(df)
    macd_chart = create_macd_chart(df)

    # 创建网格布局
    page = Page()
    page.add(rsi_chart, kdj_chart,lk_chart,macd_chart)
    # 保存为 HTML
    # grid.render(output_file)
    page.render("charts_page.html")
# 调用函数生成 HTML 文件
save_charts_to_html(df)
```
